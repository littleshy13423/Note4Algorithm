## 1.Vector
使用场景：当需要动态数组功能时，即元素数量在运行时可能变化。

优点：支持随机访问，性能良好；易于使用和理解。

缺点：在中间插入或删除元素效率低；容量增加时可能需要重新分配内存。
### 使用：
初始化vector < int > v; 为空容器

初始化vector < int > v(m);

默认值为0，即使用m个0来初始化

v.push_back(n); 在容器后方添加元素

## 2.list
使用场景：需要频繁在序列中间插入或删除元素。

优点：插入和删除操作快。

缺点：不支持直接的随机访问；比vector更多的内存开销。

## 3.deque

两头都可以快速插入的Vector

使用场景：需要在两端都能快速插入和删除的序列。

优点：两端插入和删除快；支持随机访问。

缺点：中间插入和删除效率低；内存使用不如vector高效。

## 4.map

使用场景：需要键-值对存储，且键具有唯一性。

优点：自动排序；键值快速查找。

缺点：相比于哈希表，查找效率略低。

## 5.unordered_map

使用场景：需要快速的无序键-值对存储。

优点：一般比map快；适合于需要快速查找、插入和删除的场景。

缺点：没有map那样的排序。

### 使用

初始化：

std::unordered_map < key, val > umap;

在存储各个键值对时使用的是哈希函数，默认使用 STL 标准库提供的 hash < key > 哈希函数。默认哈希函数只适用于基本数据类型（包括 string 类型），而不适用于自定义的结构体或者类。

插入新的的一组map;

umap[ < key > aNewKey ] = < val > aValue;

查找 key:

find(key), 查找以 key 为键的键值对，如果找到，则返回一个指向该键值对的正向迭代器；反之，则返回一个指向容器中最后一个键值对之后位置的迭代器（ umap.end() 方法返回的迭代器 ）。 常以 umap.find(key) == umap.end() 来判断是否存在key

## 6.set

使用场景：需要存储唯一元素，并进行高效查找。

优点：元素自动排序；高效的查找和访问。

缺点：不支持直接的元素修改。

## 7.unordered_set

使用场景：需要存储不重复的元素，且不关心元素的顺序。

优点：一般比set快。

缺点：没有排序。

## 8.stack

使用场景：需要后进先出（LIFO）的数据结构。

优点：提供快速的插入和删除。

缺点：仅限于栈顶操作，不支持随机访问。

## 9.queue

优点：
queue：简单易用，适用于FIFO操作。

缺点：
queue：仅限于队列前端和后端的操作。
