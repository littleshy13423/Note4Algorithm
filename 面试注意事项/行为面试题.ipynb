## 面试准备清单

### 遇到过的挑战

我遇到过许多挑战，但最困难的莫过于开始学习一项新的技术或是新的知识。都说贵在坚持，我承认坚持却是同等重要，但坚持的前提应该是开始。
就拿我的毕业设计来说，最困难的是两个开始。
第一个开始是开始学习骨骼点检测框架，也就是部署OPENPOSE和MIDIAPIPE，对于第一次接触这方面的我来说，配环境就花费了我不菲的时间。
学习如何取看需求清单，如何在有墙的情况去获取一些包，如何根据自己的实验设备去做一些调整以及调整之后面临的试错。
（OPENPOSE用到的主要的包：CUDA，cudnn, CMake）
后面我又尝试了几种其他的轻量化的OPENPOSE渐渐驾轻就熟。我觉得配环境是没有什么捷径的。
第二次开始是开始flutter的学习。在开始这个项目的时候我根本不会想到自己会去学习一门叫做dart的语言（我不知道当时语言流行度排多少，最近一次看倒是升到了前15）。
当时的flutter中文社区可以说还是非常不健全，我半啃着英文资料，修改一个拥有拍照功能的demo和阅读dio包的文档，完成了现在这样的功能，最终没能完成实时通信，但当时确实困难重重。

### 遇到过的滑铁卢

就比如我毕设中期决定使用云计算的方案之后，一开始定的目标是实时通信，之前没有接触过，不了解其中的难点。
还选择了一个教程不那么丰富的Flutter作为切入点，最终的结果没能完成实时通信确实是一大失策。
