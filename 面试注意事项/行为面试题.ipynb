## 面试准备清单

### 遇到过的挑战

我遇到过许多挑战，但最困难的莫过于开始学习一项新的技术或是新的知识。都说贵在坚持，我承认坚持却是同等重要，但坚持的前提应该是开始。
就拿我的毕业设计来说，最困难的是两个开始。
第一个开始是开始学习骨骼点检测框架，也就是部署OPENPOSE和MIDIAPIPE，对于第一次接触这方面的我来说，配环境就花费了我不菲的时间。
学习如何取看需求清单，如何在有墙的情况去获取一些包，如何根据自己的实验设备去做一些调整以及调整之后面临的试错。
（OPENPOSE用到的主要的包：CUDA，cudnn, CMake）
后面我又尝试了几种其他的轻量化的OPENPOSE渐渐驾轻就熟。我觉得配环境是没有什么捷径的。
第二次开始是开始flutter的学习。在开始这个项目的时候我根本不会想到自己会去学习一门叫做dart的语言（我不知道当时语言流行度排多少，最近一次看倒是升到了前15）。
当时的flutter中文社区可以说还是非常不健全，我半啃着英文资料，修改一个拥有拍照功能的demo和阅读dio包的文档，完成了现在这样的功能，最终没能完成实时通信，但当时确实困难重重。

### 遇到过的滑铁卢

就比如我毕设中期决定使用云计算的方案之后，一开始定的目标是实时通信，之前没有接触过，不了解其中的难点。
还选择了一个教程不那么丰富的Flutter作为切入点，最终的结果没能完成实时通信确实是一大失策。

### 最享受什么

我享受全身心投入某一项工作或活动，我认为能够暂时忘记其他事，花上一段连续的时间忘我地专注在手中的一件事中能让我获得很好的体验。

### 如何处理冲突

虽然我本人倾向于尽可能避免引起正面的冲突，但如果谈及冲突真正发生时，第一时间我会试图找到冲突中根本原因，
很多时候冲突的根本原因并不是表面上的一些导火索，我会尝试去找到一些新的平衡点。

### 有哪些可改进之处

和谈劣势的问题差不多

### 你最大的优势是什么

我总是乐于去尝试一些新的东西，接纳新的观点，尝试新的技术，就比如在做远程医疗项目的时候，我知道可以用 Android Studio 来做，
但当我听说有一种新的工具能够同时兼容 Android、iOS 甚至是 Window 和 Web 的时候，我会跟倾向于尝试这种新的工具。
而为本人也常常会有一些具有创新性的想法。

### 你最大的劣势是什么

当然任何一种性格或者特质都会有它的两面性，在学习的过程中，你越是深入越会发现自己的无知，
在这一过程中，我不可避免的会对新的方向产生兴趣，同时我也希望去做一些尝试的时候，它同样会分散我的注意力，让我更难以专注手边的东西。
我有注意到这一点，所以我会选择去将一些我认为好的想法记下来，而不是立刻去做这样有助于我去继续focus在原本的方向上。
同时在给自己制定计划的时候我会注意给出一些具体的，就比如读书，如果计划是一个月读完一本书，那我不一定能完成计划，
但我将计划改成每天11点到12看一个小时这本书，不用一个月，我一个礼拜就吧这本书看完了。我会通过制定一些具体的计划来克服这一劣势。
