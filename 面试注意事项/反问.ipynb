## 反问环节

最好提前准备好一到两个问题。

面试官有可能会在面试过程中谈及职位和项目相关的信息，面试者可以就这些方面的疑问进行追问。

在技术面试时可以选择对提到过的某些知识点进行追问，如果涉及算法，可以是面试官对于面试者代码的评价，以及是否有优化的建议。

注意：不要询问工资或是否能收到offer，因为大多数情况下你的面试官并不最终决定是否录取你，并且工资谈判也不是他们的工作。

### 真实的问题

1）“整个团队中，测试人员，开发人员和项目经理的比例是多少？他们如何互动？整个团队的项目规划是什么？”

2）“您当时选择该公司主要考虑的因素是什么，公司文化最吸引您的地方在哪里，您遇到过最大的挑战是什么？”

### 展现技术能力的问题

T.T 感觉需要等以后有机会社招的时候才能用得到捏

1）我有注意到你们使用了X技术，请问你们是如何处理Y问题的？

2）“为什么你们的产品选择X协议？据我所知，X虽然有A，B，C的好处，但存在D问题，很多公司并未采用该协议。

### 富有激情的问题

1）我对X技术很感兴趣，想要了解更多，有哪些机会可以学习这方面的知识？

2）我对X技术不是太熟悉，不过听上去是个不错的解决方案。您能给我多讲讲它的工作原理吗？
